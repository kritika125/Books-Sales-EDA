 # Data cleaning

```python
import pandas as pd
```

```python
df= pd.read_csv('amazon_products.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.drop([18,37,58],inplace=True)
```

```python
import re
```

```python
df["Rating"].replace("out of 5 stars", "", regex=True, inplace= True)
df
```

```python
def clean(x):
   x= re.sub(r'\s+\d*[,]*\d+',"",x)
   return x
df['Rating']= df['Rating'].apply(clean)
```

```python
df
```

```python
df["Price"].replace(r'[$]', "", regex=True, inplace= True)
df['Customers_Rated'].replace(r'[,]', "", regex=True, inplace= True)
df
```

```python
df.index = range(len(df))
```

```python
df
```

# Books with highest and lowest rating

```python
df[df.Rating== df.Rating.max()]
```

```python
df[df.Rating== df.Rating.min()]
```

```python
convert_dict = {'Price': float, 'Customers_Rated': int, 'Rating': float} 
df = df.astype(convert_dict)
print(df.dtypes)
```

# Most and least expensive book

```python
df[df.Price== df.Price.max()]
```

```python
df[df.Price== df.Price.min()]
```
